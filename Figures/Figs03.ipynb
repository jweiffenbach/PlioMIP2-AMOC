```python
folder = '/Volumes/External/DataPlioMIP2/Data/Processed/'
modellist =  ['GISS2.1G']
```

```python
##Create arrays that contain zonal mean salinity, potential temperature and potential density for E280 and Eoi400, code from Fig04FigS04.ipynb

#Use the depths from IPSL-CM6A to interpolate all other models to
model = 'IPSL-CM6A'
ds = xr.open_dataset(folder+model+'/E280/'+'so_zm_100yr.nc')
a = ds.z.values

dsE280 = xr.Dataset(data_vars = dict(rho=(["model", "z","lat"], np.zeros([len(modellist),len(ds.z), len(ds.lat)])), so=(["model", "z","lat"], np.zeros([len(modellist),len(ds.z), len(ds.lat)])), thetao=(["model", "z","lat"], np.zeros([len(modellist),len(ds.z), len(ds.lat)]))), coords = dict(model = modellist, z = ds.z.values, lat = ds.lat.values))
dsEoi400 = xr.Dataset(data_vars = dict(rho=(["model", "z", "lat"], np.zeros([len(modellist), len(ds.z), len(ds.lat)])),so=(["model", "z","lat"], np.zeros([len(modellist),len(ds.z), len(ds.lat)])),thetao=(["model", "z","lat"], np.zeros([len(modellist),len(ds.z), len(ds.lat)]))), coords = dict(model = modellist, z = ds.z.values, lat = ds.lat.values))

for i in range(len(modellist)):
    model = modellist[i]
    ds1 = xr.open_dataset(folder+model+'/E280/'+'so_zm_100yr.nc')
    ds2 = xr.open_dataset(folder+model+'/E280/'+'thetao_zm_100yr.nc')
    dsE280.so[i] = ds1.so.interp(z = a, method='linear')
    dsE280.thetao[i] = ds2.thetao.interp(z = a, method='linear')
    dsE280.rho[i] = gsw.density.rho(dsE280.so[i], dsE280.thetao[i], dsE280.so[i]*0)
    
    ds1 = xr.open_dataset(folder+model+'/Eoi400/'+'so_zm_100yr.nc')
    ds2 = xr.open_dataset(folder+model+'/Eoi400/'+'thetao_zm_100yr.nc')
    dsEoi400.so[i] = ds1.so.interp(z = a, method='linear')
    dsEoi400.thetao[i] = ds2.thetao.interp(z = a, method='linear')
    dsEoi400.rho[i] = gsw.density.rho(dsEoi400.so[i], dsEoi400.thetao[i], dsEoi400.so[i]*0)
```

```python
#Obtain the average MLD in GISS2.1G
dspi1 = xr.open_dataset('/Volumes/External/DataPlioMIP2/Data/Raw_data/'+modellist[0]+'/MLD/mlotst_Omon_GISS-E2-1-G_piControl_r1i1p1f1_gn_490101-495012.nc')
dsplio1 = xr.open_dataset('/Volumes/External/DataPlioMIP2/Data/Raw_data/'+modellist[0]+'/MLD/mlotst_Omon_GISS-E2-1-G_midPliocene-eoi400_r1i1p1f1_gn_305101-310012.nc')

dspi2 = xr.open_dataset('/Volumes/External/DataPlioMIP2/Data/Raw_data/'+modellist[0]+'/MLD/mlotst_Omon_GISS-E2-1-G_piControl_r1i1p1f1_gn_495101-500012.nc')
dsplio2 = xr.open_dataset('/Volumes/External/DataPlioMIP2/Data/Raw_data/'+modellist[0]+'/MLD/mlotst_Omon_GISS-E2-1-G_midPliocene-eoi400_r1i1p1f1_gn_310101-315012.nc')

mld_pi = dspi1.mlotst.mean(dim='time')+dspi2.mlotst.mean(dim='time')
mld_plio = dsplio1.mlotst.mean(dim='time')+dsplio2.mlotst.mean(dim='time')
```

```python
x_pos = np.array([-20,0,20,40,60,80])
xtext = ('20$\degree$S', '0$\degree$N', '20$\degree$N', '40$\degree$N', '60$\degree$N', '80$\degree$N')

cmap1 = mpl.cm.get_cmap("cmo.deep").copy()
cmap1.set_under('black')

cmap2 = mpl.cm.get_cmap("cmo.dense").copy()
cmap2.set_under('white')

plt.rcParams['xtick.labelsize']=13
plt.rcParams['ytick.labelsize']=13
plt.rcParams['axes.labelsize'] = 13

fig = plt.figure(figsize=(15,15),dpi=300)
spec = fig.add_gridspec(2, 2)

ax0 = fig.add_subplot(spec[0, 0:1])
im = ax0.pcolormesh(dsE280.lat, dsE280.z/1000, dsE280.rho[0], cmap=cmap2, shading='auto',vmin = 1022.5, vmax = 1028.5)
cp1 = ax0.contour(dsE280.lat, dsE280.z/1000, dsE280.rho[0], levels=[1026, 1027, 1027.5, 1028, 1028.25], colors='w', zorder =1, alpha=0.8)
ax0.set_xticks(x_pos)
ax0.set_xticklabels(xtext,fontsize=14)
ax0.set_xlabel('Latitude', fontsize=14)
ax0.set_ylabel('Depth (km)')
ax0.set_xlim([-34,80])
ax0.set_ylim([0,1])
ax0.set_ylim(ax0.get_ylim()[::-1])
cb = fig.colorbar(im, ax = ax0, orientation='vertical', fraction = 0.12, shrink = 0.9)
ax0.text(87,1,'(kg m$^{-3}$)', fontsize=12)
ax0.clabel(cp1, inline=True, inline_spacing=0.5, fmt = '%.1f',  fontsize=10)
ax0.set_title(r'(a)', fontsize=13,loc='left')
ax0.set_title(r'PI', fontsize=13,loc='right')

ax1 = fig.add_subplot(spec[0, 1:])

im = ax1.pcolormesh(dsEoi400.lat, dsEoi400.z/1000, dsEoi400.rho[0], cmap=cmap2, shading='auto',vmin = 1022.5, vmax = 1028.5)
cp1 = ax1.contour(dsEoi400.lat, dsEoi400.z/1000, dsEoi400.rho[0], levels=[1026, 1027, 1027.5, 1028, 1028.25], colors='w', zorder =1, alpha=0.8)
ax1.set_xticks(x_pos)
ax1.set_xticklabels(xtext,fontsize=14)
ax1.set_xlabel('Latitude', fontsize=14)
ax1.set_ylabel('Depth (km)')
ax1.set_xlim([-34,80])
ax1.set_ylim([0,1])
ax1.set_ylim(ax1.get_ylim()[::-1])
ax1.set_title(r'(b)', fontsize=13,loc='left')
ax1.set_title(r'mPWP', fontsize=13,loc='right')
cb = fig.colorbar(im, ax = ax1, orientation='vertical', fraction = 0.12, shrink = 0.9)
ax1.text(87,1,'(kg m$^{-3}$)', fontsize=12)
ax1.clabel(cp1, inline=True, inline_spacing=0.5, fmt = '%.1f',  fontsize=10)

ax2 = fig.add_subplot(spec[1, 0:1], projection=ccrs.Miller())

im = ax2.pcolormesh(mld_pi.lon, mld_pi.lat, mld_pi.fillna(-100), transform = ccrs.PlateCarree(), cmap=cmap1, shading='auto',vmin = 0, vmax = 2200)
ax2.set_title(r'(c)', fontsize=13,loc='left')
ax2.set_title(r'PI', fontsize=13,loc='right')
fig.colorbar(im, ax = ax2, location = 'bottom', fraction = 0.1, label='MLD (m)', shrink=0.8, pad=0.08)
#ax2.clabel(cp1, inline=True, inline_spacing=70, fmt = '%.1f',  fontsize=11)

ax3 = fig.add_subplot(spec[1, 1:], projection=ccrs.Miller())

im = ax3.pcolormesh(mld_plio.lon, mld_plio.lat, mld_plio.fillna(-100), transform = ccrs.PlateCarree(), cmap=cmap1, shading='auto',vmin = 0, vmax = 2200)
ax3.set_title(r'(d)', fontsize=13,loc='left')
ax3.set_title(r'mPWP', fontsize=13,loc='right')
fig.colorbar(im, ax = ax3, location = 'bottom', label='MLD (m)', fraction = 0.1, shrink=0.8, pad=0.08)
#ax3.clabel(cp1, inline=True, inline_spacing=0.1, fmt = '%.1f',  fontsize=11)

for ax in [ax2,ax3]:
    ax.set_extent([-90, 30, 20, 85], ccrs.PlateCarree())
    # Longitude labels
    ax.set_xticks(np.arange(-90,31,30), crs=ccrs.PlateCarree())
    lon_formatter = cticker.LongitudeFormatter()
    ax.xaxis.set_major_formatter(lon_formatter)

    # Latitude labels
    ax.set_yticks(np.arange(30,80,15), crs=ccrs.PlateCarree())
    lat_formatter = cticker.LatitudeFormatter()
    ax.yaxis.set_major_formatter(lat_formatter)

fig.tight_layout()
plt.savefig('figs03.png', bbox_inches='tight')
```
